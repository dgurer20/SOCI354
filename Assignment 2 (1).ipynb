```python
import pandas as pd

df_gdp_converted = df1.melt(id_vars='country', var_name='year', value_name='gdp')
df_lex_converted = df2.melt(id_vars='country', var_name='year', value_name='life_expectancy')
df_population_converted = df3.melt(id_vars='country', var_name='year', value_name='population')
df_dailyincome_converted = df4.melt(id_vars='country', var_name='year', value_name='daily_income')
df_child_mortality_converted = df5.melt(id_vars='country', var_name='year', value_name='child_mortality')

df_merged_data = pd.merge(df_gdp_converted, df_lex_converted, on=['country', 'year'])
df_merged_data = pd.merge(df_merged_data, df_population_converted, on=['country', 'year'])
df_merged_data = pd.merge(df_merged_data, df_dailyincome_converted, on=['country', 'year'])
df_merged_data = pd.merge(df_merged_data, df_child_mortality_converted, on=['country', 'year'])

print(df_merged_data)

```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_japan = df_merged_data[df_merged_data['country'] == 'Japan']


df_japan['year'] = pd.to_numeric(df_japan['year'])


df_japan = df_japan.sort_values('year')


df_japan.plot(x='year', y='Life_Expectancy', kind='line', marker='o', linestyle='-', color='purple', figsize=(9, 5))
plt.title(' Life Expectancy throughout the Years within Japan')
plt.xlabel('year')
plt.ylabel('life expectancy of Japan')
plt.grid(True)
plt.show()
```
